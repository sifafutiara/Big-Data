C. Adakah outlier pada data yang Anda miliki? Jika ada lakukan penghapusan outlier!

**Jawab:** Setelah melakukan EDA dapat disimpulkan ada outlier, sehingga outlier tersebut akan dibuang dengan menggunakan rumus IQR.

```python
#Menghitung Skor IQR
sorted(df)

Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
#Menghapus Outlier
df_clear = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df_clear
```

```python
df_clear=df_clear.select_dtypes(include=['float64', 'int64'])#Memilih kolom numerik

sns.boxplot(x="variable", y="value", color='green',orient='v', data=pd.melt(df_clear))
plt.tight_layout()
```

```python
#Menghitung Skor IQR
sorted(df1)

Q1=df1.quantile(0.25)
Q3=df1.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
#Menghapus Outlier
df_clear1 = df1[~((df1 < (Q1 - 1.5 * IQR)) |(df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
df_clear1
```

**Introduction** <br>
Random Forest adalah algoritma Supervised Machine learning yang banyak digunakan dalam masalah regresi dan klasifikasi dan sebagian besar dalam memproses data hinggan memiliki hasil ditempuh waktu yang bagus. Random Forest membangun sejumlah pohon keputusan (decision tree) pada sampel yang berbeda dan kemudian mengambil voting mayoritas (majority vote) jika memproses masalah klasifikasi.

**Kelebihan** <br>
1. Menghasilkan eror yang lebih rendah.
2. Memberikan hasil yang bagus dalam klasifikasi.
3. Dapat mengatasi data training dalam jumlah sangat besar secara efisien.
4. Metode yang efektif untuk mengestimasi hilangnya data.
5. Dapat memperkiraan variabel apa yang penting dalam klasifikasi.
6. Menyediakan metode eksperimental untuk mendeteksi interaksi variabel.

**Kekurangan** <br>
1. Waktu pemrosesan yang lama karena menggunakan data yang banyak dan membangun model tree yang banyak pula untuk membentuk random trees karena menggunakan single processor.
2. Interpretasi yang sulit dan membutuhkan mode penyetelan yang tepat untuk data.
3. Ketika digunakan untuk regresi, mereka tidak dapat memprediksi di luar kisaran dalam data percobaan, hal ini di mungkinkan data terlalu cocok dengan kumpulan data pengganggu (noisy).

**Permasalahan pada Dunia Nyata** <br>
